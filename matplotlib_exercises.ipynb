## Add an anotation for the point 0, 0, the origin.

```python
#create x datapoints
x =np.array(range(-50,50))

#create y datapoints
y = x**2 - (x+2)

#create the plot
plt.plot(x,y)

#add title and labels
plt.title('y = $x^2$ - x+2')
plt.xlabel('x axis')
plt.ylabel('y axis')

#add annotation
plt.annotate('origin', xy= (0,0), xytext = (0, 500), arrowprops= {'arrowstyle':'fancy'})

#remove banner
plt.show()
```

# Combine the figures you created in the last step into one large figure with 4 subplots.



```python
# plot different mathematical functions vs x in a single figure
x = range(21)

y1 = [math.sqrt(x) for x in x]
y2 = [x ** 3 for x in x]
y3 = [2 ** x for x in x]
y4 = [1/(x + 1) for x in x]

plt.figure(figsize = (15,11))
plt.suptitle('Some Math Functions')

plt.subplot(221)
plt.plot(x, y1)
plt.title('Lable 1) Square Root X')

plt.subplot(222)
plt.plot(x, y2)
plt.title('Lable 2) X to the power of 3')

plt.subplot(223)
plt.plot(x, y3)
plt.title('Lable 3) 2 to the power of X')

plt.subplot(224)
plt.plot(x, y4)
plt.title('Lable 4) 1 divided by x plus 1')

# alpha = controls transperancy

plt.tight_layout()

plt.show()
```

# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# plot different mathematical functions vs x in a single figure
x = range(21)

y1 = [math.sqrt(x) for x in x]
y2 = [x ** 3 for x in x]
y3 = [2 ** x for x in x]
y4 = [1/(x + 1) for x in x]

plt.figure(figsize = (8,8))
plt.suptitle('Some Math Functions')
plt.legend(loc = 'upper left')

# plt.subplot(221)
plt.plot(x, y1,color = 'r')
plt.title('Lable 1) Square Root X')

# plt.subplot(222)
plt.plot(x, y2,color = 'b')
plt.title('Lable 2) X to the power of 3')

# plt.subplot(223)
plt.plot(x, y3,color = 'g', label = '$2^x$')
plt.title('Lable 3) 2 to the power of X')

# plt.subplot(224)
plt.plot(x, y4,color = 'm')
plt.title('Lable 4) 1 divided by x plus 1')

# alpha = controls transperancy

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = range(21)

s1 = [math.sqrt(x) for x in x]
s2 = [x ** 3 for x in x]
s3 = [2 ** x for x in x]
s4 = [1/(x + 1) for x in x]

fig, axs = plt.subplots(4, 1, sharex=True)
# Remove vertical space between axes
fig.subplots_adjust(hspace=0)

# Plot each graph, and manually set the y tick values
axs[0].plot(t, s1)
axs[0].set_yticks(np.arange(0.0, 20.0, 2.5))
axs[0].set_ylim(0, 20)

axs[1].plot(t, s2)
axs[1].set_yticks(np.arange(0.0, 20.0, 2.5))
axs[1].set_ylim(0, 20)

axs[2].plot(t, s3)
axs[2].set_yticks(np.arange(0.0, 20.0, 2.5))
axs[2].set_ylim(0, 20)

axs[2].plot(t, s4)
axs[2].set_yticks(np.arange(0.0, 20.0, 2.5))
axs[2].set_ylim(0, 20)

plt.show()
```
