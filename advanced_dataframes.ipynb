# 12 Create a column named mileage_difference like you did in the DataFrames exercises; 

# this column should contain the difference between highway and city mileage for each car.



```python
# remember the sintax to create a new column 
# dataframe['name_of_column'] = information you want to display 

# this case its the differerance between two columns
# the column is now created and added to the dataset 
# no need to save or redefine the dataset
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg
```

# 13 Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.



```python
# remember the sintax to create a new column 
# dataframe['name_of_column'] = information you want to display 

# this case its the average between two columns ((a + b) / 2)
# the column is now created and added to the dataset 
# no need to save or redefine the dataset

mpg['average_mileage'] = (mpg.highway + mpg.city) / 2
mpg
```

# 14 Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.



```python
# remember the sintax to create a new column 
# dataframe['name_of_column'] = information you want to display 

# this case we are dealing with bool values and want to categorize the transmission column as true or false 
# ues np.where() that takes in a column in this case and will categorize it into on of two values
# the column is now created and added to the dataset 
# no need to save or redefine the dataset

mpg['is_automatic'] = np.where(mpg.transmission.str.startswith('a') , True, False)

mpg
```
