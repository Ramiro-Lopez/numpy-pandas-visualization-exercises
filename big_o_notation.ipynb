```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import math 
from random import randint
import random
```

# Make a new Jupyter notebook named big_o_notation.ipynb



# Curves to graph

# y = 0n + 1 and label the curve "O(1)"


```python
n = range(1, 11)
y = [0*n + 1 for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')
plt.plot(n, y, label = 'O(1)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = log(n) and label the curve "O(log n) 


```python
n = range(1, 11)
y = [math.log(n) for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(log n)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = n and label the curve "O(n)"


```python
n = range(1, 11)
y = n

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(n)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = n * log (n) and label it "O(n log n)"


```python
n = range(1, 11)
y = [n * math.log(n) for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(n log n)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = n^2 and label it "O(n^2)"


```python
n = range(1, 11)
y = [n ** 2 for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(n^2)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = 2^n and label it "O(2^n)"


```python
n = range(1, 11)
y = [2 ** n for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(2^n)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = n! and label it "O(n!)"


```python
n = range(1, 11)
y = [math.factorial(n) for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(n!)')
plt.legend(loc = 'upper left')
plt.show()
```



# y = n^n and label it "O(n^n)"

```python
n = range(1, 11)
y = [n ** n for n in n]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Opersations')

plt.plot(n, y, label = 'O(n^n)')
plt.legend(loc = 'upper left')
plt.show()
```

```python
x= (range(1,11))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4= [(n * math.log(n)) for n in x]
y5= [n**2 for n in x]
y6= [2**n for n in x]
y7= [math.factorial(n) for n in x]
y8= [n**n for n in x]

plt.figure(figsize = (16,8))

plt.plot(x,y1, color = 'cyan', label= 'y1= $O(1)$', alpha = .2)
plt.plot(x,y2, color = 'red', label= 'y2 = $O(log n)$', alpha = .3)
plt.plot(x,y3, color = 'b', label= 'y3= $O(n)$', alpha = .4)
plt.plot(x,y4, color= 'g', label= 'y4 = $O(n log n)$', alpha = .5)
plt.plot(x,y5, color= 'coral', label= 'y5 = $O(n^2)$', alpha = .6)
plt.plot(x,y6, color= 'olive', label= 'y6 = $O(2^n)$', alpha = .7)
plt.plot(x,y7, color= 'lime', label= 'y7 = $O(n!)$', alpha = .8)
plt.plot(x,y8, color= 'violet', label= 'y8 = $O(n^n)$', alpha = .9)

plt.title('Big O Notation', fontsize = 20)
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend(loc = 'upper right', fontsize = 14)
plt.xlim(1, 11)
plt.ylim(0, 50)

plt.show()
```
